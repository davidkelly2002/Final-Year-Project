```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_neurons = 10          # Number of neurons
duration = 1.0          # Duration of simulation in seconds
sampling_rate = 1000    # Sampling rate in Hz
gamma_frequency = 40    # Gamma frequency in Hz (40 Hz is common for gamma oscillations)
time = np.arange(0, duration, 1/sampling_rate)  # Time vector

# Generate neuron oscillations
def generate_oscillations(frequency, duration, sampling_rate, phase_shift=0):
    time = np.arange(0, duration, 1/sampling_rate)
    # Create a sine wave at the gamma frequency with optional phase shift
    oscillations = np.sin(2 * np.pi * frequency * time + phase_shift)
    return oscillations

# Create random phase shifts and small frequency variations for each neuron
neuronal_data = []
for i in range(n_neurons):
    phase_shift = np.random.uniform(0, 2*np.pi)  # Random phase shift for each neuron
    frequency_shift = np.random.normal(gamma_frequency, 2)  # Small variation in frequency
    oscillations = generate_oscillations(frequency_shift, duration, sampling_rate, phase_shift)
    noise = np.random.normal(0, 0.1, len(time))  # Adding some noise to the oscillations
    neuronal_data.append(oscillations + noise)

# Convert neuronal data to a NumPy array
neuronal_data = np.array(neuronal_data)

# Plot the gamma oscillations for a few neurons
plt.figure(figsize=(10, 6))
for i in range(3):  # Plot for 3 neurons
    plt.plot(time, neuronal_data[i] + i * 2, label=f'Neuron {i+1}')  # Offset each neuron for visibility
plt.title('Simulated Gamma Oscillations of Neurons')
plt.xlabel('Time (s)')
plt.ylabel('Membrane Potential (arbitrary units)')
plt.legend()
plt.show()
```
